#### ▶스쿨존 중복 정보 제거 (ex) 초등학교 부설 유치원

```python
스쿨존정보 = 스쿨존정보.drop_duplicates('소재지도로명주소', keep='first')
스쿨존정보.info()
스쿨존정보
```

#### ▶스쿨존 정보 '특수학교' 정보 정제

```python
스쿨존정보['대상시설명'] = 스쿨존정보['대상시설명'].str.replace("초등학교","초")
스쿨존정보['대상시설명'] = 스쿨존정보['대상시설명'].str.replace("초교","초")
스쿨존정보['대상시설명'] = 스쿨존정보['대상시설명'].str.replace("(특수)","")
스쿨존정보['대상시설명'] = 스쿨존정보['대상시설명'].str.replace("청주분원","")
스쿨존정보['대상시설명'] = 스쿨존정보['대상시설명'].str.replace("청주","")

스쿨존정보.tail(5)

# 청주맹학교 혜화학교(특수) 청주혜원학교 청주성신학교 충주성심학교 청주분원
```

#### ▶ 스쿨존 정보에 신호등 정보 추가

```python
light = [] 
스쿨존명 = list(스쿨존정보['대상시설명'])
# 스쿨존명 = ['한벌초']
신호등정보 = 신호등정보.dropna(subset = ['교차로'])  # 교차로 결손값 처리

for zone in 스쿨존명:
    if any(신호등정보['교차로'].str.contains(zone)):
        light.append('Y')
    else:
        light.append('N')
    
스쿨존정보['신호등'] = light
스쿨존정보

# 스쿨존정보[신호등정보['교차로'].str.contains(스쿨존정보['대상시설명'])]
```

#### ▶스쿨존 정보에 지역구 정보 추가

```python
ad = []
for i in 스쿨존정보['소재지도로명주소']:
    if i[9:12] == '서원구':
        ad.append('서원구')
    elif i[9:12] == '상당구':
        ad.append('상당구')
    elif i[9:12] == '청원구':
        ad.append('청원구') 
    else:
        ad.append('흥덕구')   
ad
```

```python
스쿨존정보['지역구'] = ad
```

#### ▶ 스쿨존정보에 교통사고 정보 넣기

```python
accident = [] 
death = []
스쿨존명 = list(스쿨존정보['대상시설명'])
지역구명 = list(스쿨존정보['지역구'])
# 스쿨존명 = ['분평초']
# 지역구명 = ['흥덕구']

스쿨명 = list(zip(스쿨존명,지역구명))
# print(스쿨명)
# 스쿨존명 = ['분평초']
# print(청주스쿨존사고[청주스쿨존사고['지점명'].str.contains('분평초')])


# FID = 청주스쿨존사고[청주스쿨존사고['지점명'].str.contains('한벌초') & 청주스쿨존사고['지점명'].str.contains('서원구')]['사고다발지FID'].tolist()
# area = 스쿨존정보[스쿨존정보['대상시설명']==zone]['지역구'].tolist()[0]
# print(FID, area)

for zone,area in 스쿨명:
#     zone = 스쿨존명[x]
#     area = 지역구명[x]
    FID = ""
#     print(zone, area)

    FID = 청주스쿨존사고[(청주스쿨존사고['지점명'].str.contains(area)) & (청주스쿨존사고['지점명'].str.contains(zone))]['index'].tolist()
#     print(FID)
    if FID:
        ac = 청주스쿨존사고[청주스쿨존사고['index']==FID[0]]['발생건수'].tolist()[0]
        de = 청주스쿨존사고[청주스쿨존사고['index']==FID[0]]['사망자수'].tolist()[0]
#         print(ac,de)
        accident.append(ac)
        death.append(de)
    else:
        accident.append(0)
        death.append(0)
스쿨존정보['발생건수'] = accident
스쿨존정보['사망자수'] = death

# 스쿨존정보[스쿨존정보['발생건수']!=0]
```
