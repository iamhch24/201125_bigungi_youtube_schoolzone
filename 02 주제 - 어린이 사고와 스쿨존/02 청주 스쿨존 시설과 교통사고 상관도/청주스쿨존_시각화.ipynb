```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
```

### 데이터 출처
- 청주스쿨존_데이터셋.ipynb

```python
스쿨존 = pd.read_csv("cj_schoolzone_Data.csv", engine = "python")
스쿨존.head(5)
```

#### ▶ 필요한 정보만 추출

```python
data = 스쿨존[['지역구', '대상시설명', '신호등', 'CCTV설치여부', '발생건수', '사망자수']]
```

#### ▶ Y/N 정보를 1, 0 으로 mapping

```python
data['신호등'] = data['신호등'].map({'Y':1, 'N':0})
data['CCTV'] = data['CCTV설치여부'].map({'Y':1, 'N':0})
```

#### ▶ 스쿨존 갯수를 위해 스쿨존이름이 있는 경우 1로 카운트

```python
data['스쿨존'] = data['대상시설명'].map(lambda x: x!="")
```

#### ▶ 지역구를 기준으로 집계

```python
data_sum = data.groupby('지역구').sum()
```

```python
data_sum.reset_index(inplace=True)
```

```python
data_sum
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x=data_sum['지역구'], y=data_sum['스쿨존'])
plt.xlabel('지역구', fontsize = 20)
plt.ylabel('스쿨존(갯)수', fontsize = 20)
plt.title("지역구별 스쿨존 개수의 관계", fontsize = 30)
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.lineplot(x='지역구', y='신호등', data=data_sum, label='신호등')
sns.lineplot(x='지역구', y='CCTV', data=data_sum, label='CCTV')
sns.lineplot(x='지역구', y='발생건수', data=data_sum, label='발생건수')
sns.lineplot(x='지역구', y='사망자수', data=data_sum, label='사망자수')
sns.lineplot(x='지역구', y='스쿨존', data=data_sum, label='스쿨존')
plt.legend(fontsize=20, loc = 'best', bbox_to_anchor=(1.3,0.7))
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('지역구', fontsize = 20)
plt.ylabel('설치된 곳/발생건수/사망자수 등등', fontsize = 20)
plt.title("지역구별 스쿨존 개수의 관계", fontsize = 30)
plt.show()
```

### ▶상관분석을 위한 스쿨존과의 비율 

```python
data_sum['신호등비율'] = round(data_sum['신호등']/data_sum['스쿨존'],2)
data_sum['CCTV비율'] = round(data_sum['CCTV']/data_sum['스쿨존'],2)
data_sum['어린이비율'] = round(data_sum['어린이인구']/data_sum['스쿨존'],2)
data_sum['사고비율'] = round(data_sum['발생건수']/data_sum['스쿨존'],2)
data_sum
```

```python
df = data_sum[['지역구','어린이비율','CCTV비율','신호등비율','사고비율']]
df
```

```python
plt.figure(figsize=(12,7))
kc = df.corr()
sns.heatmap(kc, annot=True, annot_kws={'size':20})
plt.xticks(fontsize=20)
plt.yticks(fontsize=17)
plt.title("청주 어린이,CCTV,신호등,사고 비율별 상관도", fontsize = 30)
```
