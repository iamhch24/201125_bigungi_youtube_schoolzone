#### ▶ hue에 넣기 위해 데이터 정리

```python
어린이_사망 = df.loc[:,['년도', '어린이_사망(명)']] # 년도와 어린이_사망 칼럼의 모든 값을 뽑아서 어린이_사망에 할당
어린이_사망.columns = ['년도', '건']                # 컬럼을 '년도'와 '건'으로 변경
스쿨존_교통사고 = df.loc[:, ['년도', '스쿨존_교통사고(건)']] 
스쿨존_교통사고.columns = ['년도', '건']            # 컬럼을 '년도'와'건' 변경한 이유는 두 데이터 프레임을 한 그래프에 표현하기 위해
어린이_사망['사망or교통사고'] = '어린이 사망 * 3'   
스쿨존_교통사고['사망or교통사고'] = '스쿨존 교통사고' # hue 값을 넣기 위해 새로운 칼럼안에, 값을 생성, 타이타닉 예제에서 성인 어린이로
어린이_사망['건'] = 어린이_사망['건'] * 3           # 값의 차이가 나기 때문에 *3 을 하여 아래 그래프에서 관계성을 더 직관적으로 표현하고자 하였다.
사망과사고 = 어린이_사망.append(스쿨존_교통사고)
# type(사망과사고)
# print(사망과사고)
어린이_교통사고 = df.loc[:,['년도', '어린이_교통사고(건)']]
어린이_교통사고.columns = ['년도', '값']
스쿨존_지정 = df.loc[:, ['년도', '스쿨존_지정(곳)']]
스쿨존_지정.columns = ['년도', '값']
어린이_교통사고['사고or스쿨존'] = '어린이_교통사고'
스쿨존_지정['사고or스쿨존'] = '스쿨존_지정'
사고와스쿨존 = 어린이_교통사고.append(스쿨존_지정)
```

# 이하부터 PPT에 올라갈 그래프

## 아래 그래프 해석
- 스쿨존 갯수의 증가는 어린이 교통사고를 확실히 줄여주는 효과가 있다
- 다만 최근 2019년도의 경우 예외적으로 교통사고가 증가하였다

```python
plt.figure(figsize=(12,7))
plt.plot(어린이_교통사고['년도'], 어린이_교통사고['값'], 'r', marker = 's',
        label = '어린이 교통사고 횟수', linewidth=3.0)
plt.plot(스쿨존_지정['년도'], 스쿨존_지정['값'], 'g', marker = 'o',
        label = '스쿨존 갯수', linewidth=3.0)
plt.legend(loc = 'best', fontsize = 20,)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('년도', fontsize = 20)
plt.ylabel('사고(건)과 스쿨존(갯)수', fontsize = 20)
plt.title("어린이 교통사고와 스쿨존 개수의 관계", fontsize = 30)
plt.show()
```

## ---------------------------------------------------------------------------

## 아래 그래프 설명 및 해석
- 어린이 사망 건수와 스쿨존 교통사고 건수의 시각 비교를 위해 사망 건수를 3배로 하였다
      (y축 값을 두개로 설정하는걸 수업시간에 배웠던거 같은데 기억이 가물가물)
- 두 그래프의 비교를 통해 어린이 사망숫자는 줄어 들었지만 스쿨존 교통사고 건수는
    줄어들지 않았음을 알수 있다.
- 위 그래프와 아래 그래프를 연관지어 보면 스쿨존의 증가는 어린이 사망숫자를 줄이는데 효과적이나 스쿨존 내의 교통사고를 줄이는 데는 다소 관련성이 적은 것을 알 수 있다.
- 스쿨존 내의 교통사고는 다른 직접적인 원인이 있을 것이다.

```python
# 필요한 정보만 선택
어린이_사망2 = 어린이_사망.drop(['사망or교통사고'], axis = 1)
스쿨존_교통사고2 = 스쿨존_교통사고.drop(['사망or교통사고'], axis = 1)
```

```python
plt.figure(figsize=(12,7))
plt.plot(어린이_사망2['년도'], 어린이_사망2['건'], 'r', marker = 's',
        label = '어린이 사망 건수 x 3', linewidth=3.0)
plt.plot(스쿨존_교통사고2['년도'], 스쿨존_교통사고2['건'], color = 'C1', marker = 'o',
        label = '스쿨존 교통사고 건수', linewidth=3.0)
plt.legend(loc = 'best', fontsize = 20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('년도', fontsize = 20)
plt.ylabel('사망(건)과 스쿨존 사고(건)', fontsize = 20)
plt.title("어린이 사망 건수와 스쿨존 교통사고 건수", fontsize = 30)
plt.show()
```
