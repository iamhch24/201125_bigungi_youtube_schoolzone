## 통계 검증

```python
from scipy import stats
```

```python
정치검증 = pd.read_csv('정치_검증_데이터.csv', engine = 'python')
```

```python
정치검증['구독자'] = 정치검증['구독자'] * 1000
정치검증['55세 이상 구독자 수'] = 정치검증['구독자'] * 정치검증['55세 이상 비율']
```

```python
정치검증
```

```python
보수 = 정치검증.loc[정치검증['진영'] == '보수']['55세 이상 구독자 수']
진보 = 정치검증.loc[정치검증['진영'] == '진보']['55세 이상 구독자 수']
```

### 등분산 검정

```python
lresult = stats.levene(보수, 진보)
print('LeveneResult(F) : %.3f \np-value : %.3f' % (lresult))
```

- 귀무가설: 두 집단의 분산은 같다.
- 대립가설: 두 집단의 분산은 같지 않다.
    
p-value는 0.759으로, 유의수준 0.05보다 크므로 
귀무가설이 채택되어 등분산성을 갖는 것으로 판단되었다.

등분산성을 가지므로 equal_var=True로 입력하고 
계속 독립표본 t-검정을 진행한다.

### 독립표본 t-검정

```python
result = stats.ttest_ind(보수,진보, equal_var=True) 
print('t statistic : %.3f \np-value : %.3f' % (result))
```

- 귀무가설: 두 집단의 평균은 같다.
- 대립가설: 두 집단의 평균은 같지 않다.
    
p-value가 0.03으로, 유의수준 0.05보다 작으므로 귀무가설을 기각한다.
즉, 대립가설을 채택하므로 50대 이상 연령대의 진보와 보수의 평균은 같지 않다.

그러므로 두 집단의 평균 차이가 통계적으로 유의하다고 볼 수 있다.
